```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Users/HP/OneDrive/Desktop/ongc/Pivot Table for 2023-24 SM3.xlsx")
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print(missing_values)

# Display rows with missing values
df[df.isnull().any(axis=1)]
```

```python


# Drop rows with missing values
df_cleaned = df.dropna()
```

```python
df
```

```python
# Fill missing values in daily production columns with their median
df['Daily Oil'] = df['Daily Oil'].fillna(df['Daily Oil'].median())
df['Daily Gas'] = df['Daily Gas'].fillna(df['Daily Gas'].median())
df['Daily Liquid '] = df['Daily Water'].fillna(df['Daily Water'].median())
df['Daily Total Oil+Cond'] = df['Daily Water'].fillna(df['Daily Water'].median())
df['Daily FG '] = df['Daily Water'].fillna(df['Daily Water'].median())

# Repeat for other daily columns as necessary
```

```python
df = df.dropna(subset=['Daily Oil', 'Daily Gas', 'Daily Water', 'Daily Cond', 'Daily FG'])
```

```python
df = df.drop(columns=['Layer'])
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print(missing_values)

# Display rows with missing values
df[df.isnull().any(axis=1)]
```

```python
# Find duplicates
duplicates = df[df.duplicated()]
print(duplicates)

# Remove duplicates
df_cleaned = df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt

# Plot daily oil production over time
df.groupby('Month')['Daily Oil'].sum().plot(kind='line')
plt.title('Daily Oil Production Over Time')
plt.xlabel('Month')
plt.ylabel('Daily Oil Production (SM3)')
plt.show()
```

```python
# Save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_production_data.csv', index=False)
```

```python
# Descriptive statistics of production data
production_stats = df[['Daily Oil', 'Daily Gas', 'Daily Water']].describe()
print(production_stats)
```

```python
# Correlation matrix to find relationships
correlation_matrix = df[['Daily Oil', 'Daily Gas', 'W/C', 'NW']].corr()
print(correlation_matrix)
```

```python
# Plot trends of oil, gas, and water production
df.groupby('Month').sum()[['Daily Oil', 'Daily Gas', 'Daily Water']].plot()
plt.title('Production Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Production (Units)')
plt.show()
```

```python
# Compare production across different ASSETs
df.groupby('ASSET').sum()[['Daily Oil', 'Daily Gas']].plot(kind='bar')
plt.title('Daily Oil and Gas Production by Asset')
plt.ylabel('Production (SM3/MCM)')
plt.show()
```

```python
# Top oil producing fields
top_oil_fields = df.groupby('Field')['Daily Oil'].sum().sort_values(ascending=False)
print(top_oil_fields.head(10))
```

```python
from sklearn.linear_model import LinearRegression

# Example: Predicting daily oil production based on other factors
X = df[['Daily Gas', 'Daily Water', 'W/C', 'Hrs Flown']]
y = df['Daily Oil']
model = LinearRegression()
model.fit(X, y)

# Make predictions
df['Predicted Oil'] = model.predict(X)
print(df[['Daily Oil', 'Predicted Oil']].head())
```

```python
from sklearn.cluster import KMeans

# Example: Cluster wells based on daily production
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(df[['Daily Oil', 'Daily Gas', 'Daily Water']])
print(df[['Well Id', 'Cluster']].head())
```

```python
# Save the cleaned and analyzed DataFrame
df.to_csv('final_production_analysis.csv', index=False)
```
